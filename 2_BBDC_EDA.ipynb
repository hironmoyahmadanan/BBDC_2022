```python
#Import Packges:
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Load sample file
info_file = pd.read_csv("s037t01_info.csv", index_col=[0], parse_dates=False)
info_file.head(2)
```

```python
#Remove unnamed index
info_file = info_file.reset_index(level=0)
```

```python
info_file['no_col'] = range(1, len(info_file) + 1)
```

```python
#Replacing the values : True with 1 and False with 0
info_file.loc[info_file.Mocap_Data_Exist == True,'Mo_Cap'] = 1
info_file.loc[info_file.Mocap_Data_Exist == False,'Mo_Cap'] = 0

info_file.loc[info_file.Video_Data_Exist == True,'Vi_Cap'] = 1
info_file.loc[info_file.Video_Data_Exist == False,'Vi_Cap'] = 0
```

```python
# Run from here - To be removed afterwards
info_file.to_excel('info_file.xlsx', index = False)
```

```python
x1 = info_file.no_col
y1 = info_file.Mo_Cap
y2 = info_file.Vi_Cap
```

```python
# Increase size of plot
plt.rcParams["figure.figsize"] = (20,5.5)

# Plotting the Data
plt.plot(x1, y1, label='Mocap_Data_Exist')

plt.xlabel('Frames')
plt.ylabel('Data_Exist')
plt.title("Missing head positions in Motion captured file - s037t01_info")
  
plt.plot(y1, 'o:g', linestyle='-', linewidth='0.1', color='blue')
  
plt.legend()
```

```python
# Increase size of plot
plt.rcParams["figure.figsize"] = (20,5.5)

# Plotting the Data
plt.plot(x1, y2, label='Video_Data_Exist')
  
plt.xlabel('Frames')
plt.ylabel('Data_Exist')
plt.title("Missing pixel information in Video captured file - s037t01_info")

plt.plot(y2, 'o:g', linestyle='-', linewidth='0.1', color='blue')
  
plt.legend()
```
